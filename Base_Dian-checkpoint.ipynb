```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from datetime import datetime
import os
from scipy import stats
from pathlib import Path
import glob

# Configuraci√≥n de visualizaci√≥n
warnings.filterwarnings('ignore')
plt.style.use('default')
plt.rcParams['figure.dpi'] = 100
sns.set_palette("husl")

print("‚úÖ Librer√≠as importadas correctamente")
print("üìä Configuraci√≥n de visualizaci√≥n establecida")

# ==============================================================================
# CELDA 3: EXPLORACI√ìN DE ARCHIVOS DISPONIBLES
# ==============================================================================

def explorar_archivos_disponibles(ruta_carpeta):
    """
    Explora los archivos disponibles en la carpeta de datos socioecon√≥micos
    """
    print(f"üìÇ EXPLORACI√ìN DE ARCHIVOS DISPONIBLES")
    print("-" * 45)
    
    ruta = Path(ruta_carpeta)
    
    if not ruta.exists():
        print(f"‚ùå La ruta no existe: {ruta_carpeta}")
        return None
    
    # Buscar diferentes tipos de archivos
    tipos_archivo = ['*.csv', '*.xlsx', '*.xls', '*.txt', '*.dta', '*.sav']
    archivos_encontrados = []
    
    for tipo in tipos_archivo:
        archivos = list(ruta.glob(tipo))
        archivos_encontrados.extend(archivos)
    
    if not archivos_encontrados:
        print(f"‚ùå No se encontraron archivos de datos en: {ruta_carpeta}")
        return None
    
    print(f"üìã Archivos encontrados ({len(archivos_encontrados)}):")
    archivos_info = []
    
    for i, archivo in enumerate(archivos_encontrados, 1):
        tama√±o_kb = archivo.stat().st_size / 1024
        extension = archivo.suffix.lower()
        
        info = {
            'numero': i,
            'nombre': archivo.name,
            'ruta': str(archivo),
            'extension': extension,
            'tama√±o_kb': tama√±o_kb
        }
        archivos_info.append(info)
        
        print(f"   {i:2d}. {archivo.name}")
        print(f"       Tipo: {extension} | Tama√±o: {tama√±o_kb:.1f} KB")
    
    return archivos_info

# Configurar ruta
ruta_datos_socioeconomicos = r"C:\Users\Edwin\Documents\Edwin\EAN\DataXperience\Data\Caracteristicas y composicion del hogar"

# Explorar archivos
archivos_disponibles = explorar_archivos_disponibles(ruta_datos_socioeconomicos)
```

```python
def cargar_datos_socioeconomicos(archivos_info):
    """
    Carga los datos socioecon√≥micos desde los archivos disponibles
    """
    if not archivos_info:
        print("‚ùå No hay archivos para cargar")
        return None
    
    print(f"\nüì• CARGA DE DATOS SOCIOECON√ìMICOS")
    print("-" * 40)
    
    datasets = {}
    
    for archivo_info in archivos_info:
        archivo_path = archivo_info['ruta']
        nombre = archivo_info['nombre']
        extension = archivo_info['extension']
        
        print(f"üìÑ Cargando: {nombre}")
        
        try:
            if extension == '.csv':
                # Intentar diferentes encodings
                encodings = ['utf-8', 'latin-1', 'cp1252', 'iso-8859-1']
                df = None
                
                for encoding in encodings:
                    try:
                        df = pd.read_csv(archivo_path, encoding=encoding)
                        print(f"   ‚úÖ Cargado con encoding: {encoding}")
                        break
                    except UnicodeDecodeError:
                        continue
                
                if df is None:
                    print(f"   ‚ùå No se pudo cargar con ning√∫n encoding")
                    continue
                    
            elif extension in ['.xlsx', '.xls']:
                df = pd.read_excel(archivo_path)
                print(f"   ‚úÖ Archivo Excel cargado")
                
            elif extension == '.dta':
                # Archivo Stata
                df = pd.read_stata(archivo_path)
                print(f"   ‚úÖ Archivo Stata cargado")
                
            elif extension == '.sav':
                # Archivo SPSS
                try:
                    import pyreadstat
                    df, meta = pyreadstat.read_sav(archivo_path)
                    print(f"   ‚úÖ Archivo SPSS cargado")
                except ImportError:
                    print(f"   ‚ö†Ô∏è pyreadstat no disponible para archivos .sav")
                    continue
                    
            else:
                print(f"   ‚ö†Ô∏è Tipo de archivo no soportado: {extension}")
                continue
            
            # Informaci√≥n b√°sica del dataset
            print(f"   üìä Dimensiones: {df.shape}")
            print(f"   üìã Columnas: {len(df.columns)}")
            
            # Guardar en diccionario
            datasets[nombre] = df
            
        except Exception as e:
            print(f"   ‚ùå Error cargando {nombre}: {e}")
    
    return datasets

# Cargar datos
if archivos_disponibles:
    datasets_socioeconomicos = cargar_datos_socioeconomicos(archivos_disponibles)
else:
    datasets_socioeconomicos = None
```

```python
def explorar_datasets_inicial(datasets):
    """
    Exploraci√≥n inicial de los datasets cargados
    """
    if not datasets:
        print("‚ùå No hay datasets para explorar")
        return
    
    print(f"\nüîç EXPLORACI√ìN INICIAL DE DATASETS")
    print("=" * 45)
    
    for nombre, df in datasets.items():
        print(f"\nüìä DATASET: {nombre}")
        print("-" * 30)
        
        # Informaci√≥n b√°sica
        print(f"Dimensiones: {df.shape}")
        print(f"Memoria utilizada: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
        
        # Tipos de datos
        print(f"\nTipos de datos:")
        tipos = df.dtypes.value_counts()
        for tipo, count in tipos.items():
            print(f"   {tipo}: {count} columnas")
        
        # Columnas disponibles (primeras 20)
        print(f"\nPrimeras 20 columnas:")
        for i, col in enumerate(df.columns[:20], 1):
            print(f"   {i:2d}. {col}")
        
        if len(df.columns) > 20:
            print(f"   ... y {len(df.columns) - 20} columnas m√°s")
        
        # Muestra de datos
        print(f"\nPrimeras 3 filas:")
        display(df.head(3))
        
        # Buscar columnas geogr√°ficas
        columnas_geograficas = []
        palabras_geo = ['municipio', 'departamento', 'ciudad', 'region', 'codigo', 'divipola']
        
        for col in df.columns:
            if any(palabra in col.lower() for palabra in palabras_geo):
                columnas_geograficas.append(col)
        
        if columnas_geograficas:
            print(f"\nüó∫Ô∏è Columnas geogr√°ficas identificadas:")
            for col in columnas_geograficas:
                print(f"   ‚Ä¢ {col}")
        
        print("-" * 50)

# Explorar datasets
if datasets_socioeconomicos:
    explorar_datasets_inicial(datasets_socioeconomicos)
```

```python
def seleccionar_dataset_principal(datasets):
    """
    Selecciona el dataset principal para an√°lisis
    """
    if not datasets:
        print("‚ùå No hay datasets disponibles")
        return None, None
    
    print(f"\nüéØ SELECCI√ìN DEL DATASET PRINCIPAL")
    print("-" * 40)
    
    # Si hay un solo dataset, usarlo
    if len(datasets) == 1:
        nombre = list(datasets.keys())[0]
        df = datasets[nombre]
        print(f"‚úÖ Usando √∫nico dataset disponible: {nombre}")
        return df, nombre
    
    # Si hay m√∫ltiples, mostrar opciones
    print(f"üìã Datasets disponibles:")
    for i, (nombre, df) in enumerate(datasets.items(), 1):
        print(f"   {i}. {nombre} - {df.shape[0]:,} filas, {df.shape[1]} columnas")
    
    # Por defecto, usar el m√°s grande
    dataset_principal = max(datasets.items(), key=lambda x: x[1].shape[0])
    nombre_principal = dataset_principal[0]
    df_principal = dataset_principal[1]
    
    print(f"\n‚úÖ Seleccionando dataset con m√°s registros: {nombre_principal}")
    print(f"   Dimensiones: {df_principal.shape}")
    
    return df_principal, nombre_principal

# Seleccionar dataset principal
if datasets_socioeconomicos:
    df_socioeconomico_raw, nombre_dataset = seleccionar_dataset_principal(datasets_socioeconomicos)
else:
    df_socioeconomico_raw, nombre_dataset = None, None
```

```python
def seleccionar_dataset_principal_corregido(datasets):
    """
    Selecciona el dataset principal para an√°lisis - VERSI√ìN CORREGIDA
    Prioriza por n√∫mero de registros Y n√∫mero de columnas
    """
    if not datasets:
        print("‚ùå No hay datasets disponibles")
        return None, None
    
    print(f"\nüéØ SELECCI√ìN DEL DATASET PRINCIPAL - VERSI√ìN CORREGIDA")
    print("-" * 55)
    
    # Mostrar informaci√≥n detallada de todos los datasets
    print(f"üìã Datasets disponibles:")
    datasets_info = []
    
    for nombre, df in datasets.items():
        filas, columnas = df.shape
        
        # Calcular score considerando tanto filas como columnas
        # Penalizar datasets con muy pocas columnas
        if columnas < 5:
            score = filas * columnas * 0.1  # Penalizaci√≥n fuerte
        else:
            score = filas * columnas
        
        info = {
            'nombre': nombre,
            'dataframe': df,
            'filas': filas,
            'columnas': columnas,
            'score': score
        }
        datasets_info.append(info)
        
        print(f"\n   üìä {nombre}")
        print(f"      üìà Filas: {filas:,}")
        print(f"      üìã Columnas: {columnas:,}")
        print(f"      üéØ Score: {score:,.0f}")
        
        # Mostrar algunas columnas como muestra
        if columnas > 0:
            print(f"      üìù Primeras columnas: {list(df.columns[:3])}")
    
    # Si hay un solo dataset, usarlo (pero con advertencia si tiene pocas columnas)
    if len(datasets) == 1:
        dataset_info = datasets_info[0]
        print(f"\n‚úÖ Usando √∫nico dataset disponible: {dataset_info['nombre']}")
        
        if dataset_info['columnas'] < 5:
            print(f"‚ö†Ô∏è ADVERTENCIA: Este dataset tiene pocas columnas ({dataset_info['columnas']})")
            print(f"   Puede no ser adecuado para an√°lisis socioecon√≥mico completo")
        
        return dataset_info['dataframe'], dataset_info['nombre']
    
    # Seleccionar el dataset con mayor score
    dataset_seleccionado = max(datasets_info, key=lambda x: x['score'])
    
    print(f"\nüèÜ DATASET SELECCIONADO:")
    print(f"   üìä Nombre: {dataset_seleccionado['nombre']}")
    print(f"   üìà Dimensiones: {dataset_seleccionado['filas']:,} filas √ó {dataset_seleccionado['columnas']:,} columnas")
    print(f"   üéØ Score final: {dataset_seleccionado['score']:,.0f}")
    
    # Raz√≥n de la selecci√≥n
    print(f"\nüí° Raz√≥n de selecci√≥n:")
    if dataset_seleccionado['columnas'] >= 10:
        print(f"   ‚úÖ Dataset con mayor cantidad de datos y columnas adecuadas para an√°lisis")
    else:
        print(f"   ‚ö†Ô∏è Dataset seleccionado por defecto, pero verificar calidad de columnas")
    
    return dataset_seleccionado['dataframe'], dataset_seleccionado['nombre']

# LLAMADA CORREGIDA DE LA FUNCI√ìN:
if datasets_socioeconomicos:
    df_socioeconomico_raw, nombre_dataset = seleccionar_dataset_principal_corregido(datasets_socioeconomicos)
    
    # Verificaci√≥n adicional
    if df_socioeconomico_raw is not None:
        print(f"\n‚úÖ DATASET CARGADO EXITOSAMENTE:")
        print(f"   Nombre: {nombre_dataset}")
        print(f"   Dimensiones finales: {df_socioeconomico_raw.shape}")
        print(f"   Memoria: {df_socioeconomico_raw.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
        
        # Mostrar primeras columnas
        print(f"\nüìã Primeras 10 columnas del dataset seleccionado:")
        for i, col in enumerate(df_socioeconomico_raw.columns[:10], 1):
            print(f"   {i:2d}. {col}")
        
        if len(df_socioeconomico_raw.columns) > 10:
            print(f"   ... y {len(df_socioeconomico_raw.columns) - 10} columnas m√°s")
    else:
        print("‚ùå No se pudo cargar ning√∫n dataset")
else:
    df_socioeconomico_raw, nombre_dataset = None, None
    print("‚ùå No hay datasets disponibles para seleccionar")
```

```python
def analisis_datos_faltantes_socioeconomicos(df):
    """
    An√°lisis exhaustivo de datos faltantes para datos socioecon√≥micos
    """
    print(f"\nüìä AN√ÅLISIS DETALLADO DE DATOS FALTANTES")
    print("=" * 50)
    
    # Conteo de nulos por columna
    nulos_por_columna = df.isnull().sum()
    porcentaje_nulos = (nulos_por_columna / len(df) * 100).round(2)
    
    # Crear DataFrame de resumen
    resumen_nulos = pd.DataFrame({
        'Columna': nulos_por_columna.index,
        'Valores_Nulos': nulos_por_columna.values,
        'Porcentaje_Nulos': porcentaje_nulos.values,
        'Valores_Completos': len(df) - nulos_por_columna.values
    }).sort_values('Porcentaje_Nulos', ascending=False)
    
    print("üìã Resumen de datos faltantes:")
    display(resumen_nulos.head(20))  # Mostrar top 20
    
    # Visualizaci√≥n de datos faltantes
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Gr√°fico de barras de porcentaje de nulos
    columnas_con_nulos = resumen_nulos[resumen_nulos['Porcentaje_Nulos'] > 0]
    if len(columnas_con_nulos) > 0:
        top_nulos = columnas_con_nulos.head(15)  # Top 15 con m√°s nulos
        ax1.barh(range(len(top_nulos)), top_nulos['Porcentaje_Nulos'])
        ax1.set_yticks(range(len(top_nulos)))
        ax1.set_yticklabels(top_nulos['Columna'], fontsize=8)
        ax1.set_title('Top 15 Columnas con M√°s Datos Faltantes')
        ax1.set_xlabel('Porcentaje (%)')
    else:
        ax1.text(0.5, 0.5, 'No hay datos faltantes', ha='center', va='center', transform=ax1.transAxes)
        ax1.set_title('Sin Datos Faltantes')
    
    # Heatmap de patrones de nulos (muestra)
    if df.isnull().sum().sum() > 0:
        # Tomar muestra para visualizaci√≥n
        muestra_size = min(1000, len(df))
        muestra_cols = min(50, len(df.columns))  # M√°ximo 50 columnas
        
        df_muestra = df.iloc[:muestra_size, :muestra_cols]
        sns.heatmap(df_muestra.isnull(), cbar=True, yticklabels=False, 
                   xticklabels=False, cmap='viridis', ax=ax2)
        ax2.set_title(f'Patr√≥n de Datos Faltantes\n(muestra {muestra_size} filas, {muestra_cols} cols)')
    else:
        ax2.text(0.5, 0.5, 'Dataset Completo', ha='center', va='center', transform=ax2.transAxes)
        ax2.set_title('Sin Datos Faltantes')
    
    plt.tight_layout()
    plt.show()
    
    # Estad√≠sticas generales
    total_columnas = len(df.columns)
    columnas_con_nulos_count = len(columnas_con_nulos)
    columnas_completas = total_columnas - columnas_con_nulos_count
    
    print(f"\nüìà ESTAD√çSTICAS GENERALES:")
    print(f"   Total de columnas: {total_columnas}")
    print(f"   Columnas completas: {columnas_completas} ({(columnas_completas/total_columnas*100):.1f}%)")
    print(f"   Columnas con nulos: {columnas_con_nulos_count} ({(columnas_con_nulos_count/total_columnas*100):.1f}%)")
    
    return resumen_nulos

# LLAMADA DE LA FUNCI√ìN:
if df_socioeconomico_raw is not None:
    resumen_nulos_socio = analisis_datos_faltantes_socioeconomicos(df_socioeconomico_raw)
```

```python
def identificar_variables_socioeconomicas(df):
    """
    Identifica las variables socioecon√≥micas clave del dataset
    """
    print(f"\nüîç IDENTIFICACI√ìN DE VARIABLES SOCIOECON√ìMICAS CLAVE")
    print("-" * 55)
    
    variables_identificadas = {
        'geograficas': [],
        'demograficas': [],
        'economicas': [],
        'vivienda': [],
        'educacion': [],
        'salud': [],
        'tecnologia': [],
        'bienes': []
    }
    
    # Palabras clave por categor√≠a
    palabras_clave = {
        'geograficas': ['municipio', 'departamento', 'ciudad', 'region', 'codigo', 'divipola', 'dane'],
        'demograficas': ['edad', 'sexo', 'poblacion', 'habitantes', 'personas', 'genero', 'ninos', 'adultos'],
        'economicas': ['ingreso', 'salario', 'empleo', 'trabajo', 'ocupacion', 'desempleo', 'pobreza', 'riqueza'],
        'vivienda': ['vivienda', 'hogar', 'casa', 'apartamento', 'cuarto', 'dormitorio', 'bano', 'cocina'],
        'educacion': ['educacion', 'escuela', 'colegio', 'universidad', 'estudio', 'alfabetismo', 'analfabeta'],
        'salud': ['salud', 'medico', 'hospital', 'enfermedad', 'medicina', 'seguridad_social'],
        'tecnologia': ['internet', 'computador', 'celular', 'telefono', 'tecnologia', 'digital'],
        'bienes': ['television', 'nevera', 'refrigerador', 'lavadora', 'carro', 'moto', 'bicicleta', 'electrodomestico']
    }
    
    # Clasificar columnas
    for col in df.columns:
        col_lower = col.lower()
        clasificada = False
        
        for categoria, palabras in palabras_clave.items():
            if any(palabra in col_lower for palabra in palabras):
                variables_identificadas[categoria].append(col)
                clasificada = True
                break
        
        # Si no se clasific√≥, intentar por tipo de dato
        if not clasificada:
            if df[col].dtype in ['int64', 'float64']:
                # Podr√≠a ser una variable num√©rica socioecon√≥mica
                if 'num' in col_lower or 'cant' in col_lower or 'total' in col_lower:
                    variables_identificadas['demograficas'].append(col)
    
    # Mostrar resultados
    print(f"üìã Variables identificadas por categor√≠a:")
    total_clasificadas = 0
    
    for categoria, variables in variables_identificadas.items():
        if variables:
            print(f"\nüè∑Ô∏è {categoria.upper()} ({len(variables)} variables):")
            for var in variables[:10]:  # Mostrar m√°ximo 10 por categor√≠a
                print(f"   ‚Ä¢ {var}")
            if len(variables) > 10:
                print(f"   ... y {len(variables) - 10} m√°s")
            total_clasificadas += len(variables)
    
    total_columnas = len(df.columns)
    sin_clasificar = total_columnas - total_clasificadas
    
    print(f"\nüìä RESUMEN DE CLASIFICACI√ìN:")
    print(f"   Total de columnas: {total_columnas}")
    print(f"   Variables clasificadas: {total_clasificadas} ({(total_clasificadas/total_columnas*100):.1f}%)")
    print(f"   Sin clasificar: {sin_clasificar} ({(sin_clasificar/total_columnas*100):.1f}%)")
    
    return variables_identificadas

# LLAMADA DE LA FUNCI√ìN:
if df_socioeconomico_raw is not None:
    variables_socioeconomicas = identificar_variables_socioeconomicas(df_socioeconomico_raw)
```

```python
def limpiar_datos_socioeconomicos(df, variables_identificadas):
    """
    Limpia y procesa los datos socioecon√≥micos
    """
    print(f"\nüßπ LIMPIEZA Y PROCESAMIENTO DE DATOS SOCIOECON√ìMICOS")
    print("-" * 55)
    
    df_clean = df.copy()
    
    # 1. Limpiar variables geogr√°ficas
    print("üó∫Ô∏è Procesando variables geogr√°ficas...")
    
    for col in variables_identificadas['geograficas']:
        if col in df_clean.columns:
            if 'municipio' in col.lower():
                df_clean[f'{col}_limpio'] = (df_clean[col]
                                           .astype(str)
                                           .str.upper()
                                           .str.strip()
                                           .str.replace('√Å', 'A')
                                           .str.replace('√â', 'E')
                                           .str.replace('√ç', 'I')
                                           .str.replace('√ì', 'O')
                                           .str.replace('√ö', 'U')
                                           .str.replace('√ë', 'N'))
                print(f"   ‚úÖ {col} -> {col}_limpio")
            
            elif 'departamento' in col.lower():
                df_clean[f'{col}_limpio'] = (df_clean[col]
                                           .astype(str)
                                           .str.upper()
                                           .str.strip())
                print(f"   ‚úÖ {col} -> {col}_limpio")
    
    # 2. Convertir variables num√©ricas
    print(f"\nüî¢ Procesando variables num√©ricas...")
    
    variables_numericas = (variables_identificadas['demograficas'] + 
                          variables_identificadas['economicas'] + 
                          variables_identificadas['bienes'])
    
    for col in variables_numericas:
        if col in df_clean.columns:
            # Intentar convertir a num√©rico
            original_type = df_clean[col].dtype
            df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
            
            valores_validos = df_clean[col].notna().sum()
            total = len(df_clean)
            porcentaje = (valores_validos / total) * 100
            
            if porcentaje > 50:  # Solo reportar si m√°s del 50% son v√°lidos
                print(f"   ‚úÖ {col}: {porcentaje:.1f}% valores num√©ricos v√°lidos")
    
    # 3. Crear variables derivadas
    print(f"\nüìä Creando variables derivadas...")
    
    # √çndice de bienes (si hay variables de bienes)
    if variables_identificadas['bienes']:
        bienes_cols = [col for col in variables_identificadas['bienes'] 
                      if col in df_clean.columns and df_clean[col].dtype in ['int64', 'float64']]
        
        if bienes_cols:
            df_clean['indice_bienes'] = df_clean[bienes_cols].sum(axis=1, skipna=True)
            df_clean['cantidad_bienes'] = df_clean[bienes_cols].notna().sum(axis=1)
            print(f"   ‚úÖ √çndice de bienes creado ({len(bienes_cols)} variables)")
    
    # √çndice socioecon√≥mico general (variables econ√≥micas)
    if variables_identificadas['economicas']:
        econ_cols = [col for col in variables_identificadas['economicas'] 
                    if col in df_clean.columns and df_clean[col].dtype in ['int64', 'float64']]
        
        if econ_cols:
            # Estandarizar y sumar
            econ_data = df_clean[econ_cols]
            econ_std = (econ_data - econ_data.mean()) / econ_data.std()
            df_clean['indice_economico'] = econ_std.sum(axis=1, skipna=True)
            print(f"   ‚úÖ √çndice econ√≥mico creado ({len(econ_cols)} variables)")
    
    print(f"\n‚úÖ Datos procesados exitosamente")
    print(f"   Registros finales: {len(df_clean):,}")
    print(f"   Columnas finales: {len(df_clean.columns)}")
    
    return df_clean

# LLAMADA DE LA FUNCI√ìN:
if df_socioeconomico_raw is not None and 'variables_socioeconomicas' in locals():
    df_socioeconomico_clean = limpiar_datos_socioeconomicos(df_socioeconomico_raw, variables_socioeconomicas)
    
    print(f"\nüìã Muestra de datos limpios:")
    display(df_socioeconomico_clean.head())
```

```python
def detectar_outliers_socioeconomicos(df, variables_numericas):
    """
    Detecta outliers en las variables socioecon√≥micas
    """
    print(f"\nüìê DETECCI√ìN DE OUTLIERS - VARIABLES SOCIOECON√ìMICAS")
    print("-" * 55)
    
    if not variables_numericas:
        print("‚ùå No hay variables num√©ricas para analizar")
        return None
    
    outliers_info = {}
    
    # Analizar las primeras 8 variables para no saturar
    vars_analizar = variables_numericas[:8]
    
    for col in vars_analizar:
        if col in df.columns:
            valores = df[col].dropna()
            
            if len(valores) == 0:
                continue
            
            print(f"\nüìä AN√ÅLISIS DE: {col.upper()}")
            print("-" * 30)
            
            # M√©todo IQR
            Q1 = valores.quantile(0.25)
            Q3 = valores.quantile(0.75)
            IQR = Q3 - Q1
            limite_inferior = Q1 - 1.5 * IQR
            limite_superior = Q3 + 1.5 * IQR
            
            outliers_iqr = valores[(valores < limite_inferior) | (valores > limite_superior)]
            
            # M√©todo Z-score
            if valores.std() > 0:
                z_scores = np.abs((valores - valores.mean()) / valores.std())
                outliers_zscore = valores[z_scores > 3]
            else:
                outliers_zscore = pd.Series(dtype=float)
            
            print(f"Estad√≠sticas b√°sicas:")
            print(f"   Media: {valores.mean():.4f}")
            print(f"   Mediana: {valores.median():.4f}")
            print(f"   Desv. est√°ndar: {valores.std():.4f}")
            print(f"   Rango: {valores.min():.4f} - {valores.max():.4f}")
            
            print(f"\nOutliers detectados:")
            print(f"   M√©todo IQR: {len(outliers_iqr):,} outliers")
            print(f"   M√©todo Z-score: {len(outliers_zscore):,} outliers")
            
            outliers_info[col] = {
                'outliers_iqr': outliers_iqr,
                'outliers_zscore': outliers_zscore,
                'limites_iqr': (limite_inferior, limite_superior)
            }
    
    # Visualizaci√≥n de outliers
    if len(vars_analizar) > 0:
        n_cols = min(len(vars_analizar), 4)
        n_rows = 2
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 10))
        
        if n_cols == 1:
            axes = axes.reshape(-1, 1)
        elif len(vars_analizar) == 1:
            axes = [[axes[0]], [axes[1]]]
        
        for i, col in enumerate(vars_analizar[:n_cols]):
            if col in df.columns:
                valores = df[col].dropna()
                
                if len(valores) > 0:
                    # Histograma
                    axes[0][i].hist(valores, bins=30, edgecolor='black', alpha=0.7)
                    axes[0][i].set_title(f'Distribuci√≥n - {col[:20]}...')
                    axes[0][i].set_xlabel('Valor')
                    
                    # Boxplot
                    axes[1][i].boxplot(valores)
                    axes[1][i].set_title(f'Boxplot - {col[:20]}...')
        
        plt.tight_layout()
        plt.show()
    
    return outliers_info

# LLAMADA DE LA FUNCI√ìN:
if 'vars_numericas_principales' in locals() and 'df_socioeconomico_clean' in locals():
    outliers_info_socio = detectar_outliers_socioeconomicos(df_socioeconomico_clean, vars_numericas_principales)
```

```python
def preparar_para_integracion_dane_socio_corregida(df, variables_identificadas):
    """
    Prepara los datos socioecon√≥micos para integraci√≥n con datos del DANE - CORREGIDA
    """
    print(f"\nüîó PREPARACI√ìN PARA INTEGRACI√ìN CON DANE - DATOS SOCIOECON√ìMICOS")
    print("-" * 65)
    
    # Identificar columna de municipio
    columna_municipio = None
    for col in variables_identificadas['geograficas']:
        if 'municipio' in col.lower() and col in df.columns:
            columna_municipio = col
            break
    
    if columna_municipio is None:
        print("‚ùå No se encontr√≥ columna de municipio")
        print("üìã Columnas geogr√°ficas disponibles:")
        for col in variables_identificadas['geograficas']:
            if col in df.columns:
                print(f"   ‚Ä¢ {col}")
        return None
    
    print(f"‚úÖ Usando columna de municipio: {columna_municipio}")
    
    # Identificar variables para agregar
    variables_a_agregar = []
    
    # Variables num√©ricas de cada categor√≠a
    for categoria in ['demograficas', 'economicas', 'bienes', 'vivienda', 'educacion', 'salud', 'tecnologia']:
        for col in variables_identificadas[categoria]:
            if col in df.columns and df[col].dtype in ['int64', 'float64']:
                variables_a_agregar.append(col)
    
    # Variables derivadas
    vars_derivadas = ['indice_bienes', 'cantidad_bienes', 'indice_economico']
    for var in vars_derivadas:
        if var in df.columns:
            variables_a_agregar.append(var)
    
    if len(variables_a_agregar) == 0:
        print("‚ùå No se encontraron variables num√©ricas para agregar")
        return None
    
    print(f"üìã Variables a agregar: {len(variables_a_agregar)}")
    for categoria in ['demograficas', 'economicas', 'bienes', 'vivienda', 'educacion', 'salud', 'tecnologia']:
        vars_categoria = [var for var in variables_a_agregar if var in variables_identificadas[categoria]]
        if vars_categoria:
            print(f"   {categoria.capitalize()}: {len(vars_categoria)} variables")
    
    # Agregar por municipio usando promedio
    print(f"\nüìä Agregando datos por municipio...")
    df_agregado = df.groupby(columna_municipio)[variables_a_agregar].mean().reset_index()
    
    # Renombrar columna de municipio para consistencia
    df_agregado = df_agregado.rename(columns={columna_municipio: 'municipio'})
    
    # Limpiar nombres de municipios para matching
    df_agregado['municipio_para_merge'] = (df_agregado['municipio']
                                          .astype(str)
                                          .str.upper()
                                          .str.strip()
                                          .str.replace('√Å', 'A')
                                          .str.replace('√â', 'E')
                                          .str.replace('√ç', 'I')
                                          .str.replace('√ì', 'O')
                                          .str.replace('√ö', 'U')
                                          .str.replace('√ë', 'N'))
    
    # Estad√≠sticas del agregado
    print(f"\nüìà Resultados de la agregaci√≥n:")
    print(f"   Municipios √∫nicos: {len(df_agregado):,}")
    print(f"   Variables socioecon√≥micas: {len(variables_a_agregar)}")
    
    # Verificar completitud
    print(f"\nüîç Verificaci√≥n de completitud (top 10 variables):")
    for col in variables_a_agregar[:10]:
        valores_validos = df_agregado[col].notna().sum()
        porcentaje = (valores_validos / len(df_agregado)) * 100
        print(f"   {col[:30]}: {porcentaje:.1f}%")
    
    if len(variables_a_agregar) > 10:
        print(f"   ... y {len(variables_a_agregar) - 10} variables m√°s")
    
    return df_agregado

# LLAMADA DE LA FUNCI√ìN:
if 'df_socioeconomico_clean' in locals() and 'variables_socioeconomicas' in locals():
    df_socio_para_merge = preparar_para_integracion_dane_socio_corregida(df_socioeconomico_clean, variables_socioeconomicas)
    
    if df_socio_para_merge is not None:
        print(f"\nüìã Muestra de datos preparados para merge:")
        display(df_socio_para_merge.head(10))
```

```python
def validar_calidad_socioeconomicos_corregida(df_original, df_procesado, df_agregado=None):
    """
    Validaci√≥n de calidad de los datos socioecon√≥micos procesados - CORREGIDA
    """
    print(f"\n‚úÖ VALIDACI√ìN DE CALIDAD - DATOS SOCIOECON√ìMICOS")
    print("-" * 50)
    
    # 1. Comparaci√≥n de registros
    print(f"üìä Comparaci√≥n de registros:")
    print(f"   Datos originales: {len(df_original):,} registros")
    print(f"   Datos procesados: {len(df_procesado):,} registros")
    
    if df_agregado is not None:
        print(f"   Datos agregados por municipio: {len(df_agregado):,} registros")
        reduccion = len(df_original) - len(df_agregado)
        print(f"   Reducci√≥n por agregaci√≥n: {reduccion:,} registros")
    
    # 2. Verificar variables num√©ricas
    vars_numericas = df_procesado.select_dtypes(include=[np.number]).columns
    print(f"\nüìä Validaci√≥n de variables num√©ricas:")
    print(f"   Variables num√©ricas identificadas: {len(vars_numericas)}")
    
    if len(vars_numericas) > 0:
        print(f"\n   Calidad de datos num√©ricos (top 10):")
        for col in vars_numericas[:10]:
            valores_validos = df_procesado[col].notna().sum()
            total = len(df_procesado)
            completitud = (valores_validos / total) * 100
            
            if len(df_procesado[col].dropna()) > 0:
                valores = df_procesado[col].dropna()
                valores_negativos = (valores < 0).sum()
                print(f"   {col[:30]:30}: {completitud:5.1f}% completo, {valores_negativos:,} negativos")
    
    # 3. Verificar duplicados en datos agregados
    if df_agregado is not None and 'municipio' in df_agregado.columns:
        duplicados = df_agregado['municipio'].duplicated().sum()
        print(f"\nüîç Verificaci√≥n de duplicados:")
        print(f"   Municipios duplicados: {duplicados}")
        
        if duplicados > 0:
            print("   ‚ö†Ô∏è ADVERTENCIA: Se encontraron municipios duplicados")
    
    # 4. Score de calidad general
    print(f"\nüìà SCORE DE CALIDAD GENERAL:")
    
    scores = []
    
    # Score de completitud
    if len(vars_numericas) > 0:
        completitud_promedio = 0
        for col in vars_numericas[:20]:  # Evaluar m√°ximo 20 variables
            valores_validos = df_procesado[col].notna().sum()
            completitud = (valores_validos / len(df_procesado)) * 100
            completitud_promedio += completitud
        completitud_promedio /= min(len(vars_numericas), 20)
        scores.append(completitud_promedio)
        print(f"   Completitud promedio: {completitud_promedio:.1f}%")
    
    # Score de consistencia
    if df_agregado is not None:
        duplicados = df_agregado['municipio'].duplicated().sum() if 'municipio' in df_agregado.columns else 0
        score_duplicados = 100 if duplicados == 0 else max(0, 100 - (duplicados * 10))
        scores.append(score_duplicados)
        print(f"   Score de unicidad: {score_duplicados:.1f}%")
    
    # Score general
    if scores:
        score_general = sum(scores) / len(scores)
        print(f"   üìä SCORE GENERAL: {score_general:.1f}%")
        
        if score_general >= 90:
            print("   üéØ CALIDAD: EXCELENTE")
        elif score_general >= 75:
            print("   ‚úÖ CALIDAD: BUENA")
        elif score_general >= 60:
            print("   ‚ö†Ô∏è CALIDAD: ACEPTABLE")
        else:
            print("   ‚ùå CALIDAD: REQUIERE MEJORAS")
    
    return True

# LLAMADA DE LA FUNCI√ìN:
if all(var in locals() for var in ['df_socioeconomico_raw', 'df_socioeconomico_clean']):
    if 'df_socio_para_merge' in locals() and df_socio_para_merge is not None:
        validacion_socio_ok = validar_calidad_socioeconomicos_corregida(df_socioeconomico_raw, df_socioeconomico_clean, df_socio_para_merge)
    else:
        validacion_socio_ok = validar_calidad_socioeconomicos_corregida(df_socioeconomico_raw, df_socioeconomico_clean, None)

validacion_socio_ok = validar_calidad_socioeconomicos_corregida(df_socioeconomico_raw, df_socioeconomico_clean, df_socio_para_merge)
```
