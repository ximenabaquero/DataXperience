```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from datetime import datetime
import requests
import json
import os
from scipy import stats
from pathlib import Path

# Configuraci√≥n de visualizaci√≥n
warnings.filterwarnings('ignore')
plt.style.use('default')
plt.rcParams['figure.dpi'] = 100
sns.set_palette("husl")

print("‚úÖ Librer√≠as importadas correctamente")
print("üìä Configuraci√≥n de visualizaci√≥n establecida")
```

```python
def descargar_posdata():
    """
    Descarga TODOS los registros de cobertura m√≥vil usando paginaci√≥n
    """
    base_url = "https://www.postdata.gov.co/api/action/datastore/search.json"
    resource_id = "77e64c24-8693-4633-875f-f70f72ff9f7a"

    all_records = []
    limit = 80000  # Registros por p√°gina
    offset = 0

    print("üì° Descargando todos los registros de cobertura m√≥vil...")

    while True:
        params = {
            'resource_id': resource_id,
            'limit': limit,
            'offset': offset
        }

        try:
            response = requests.get(base_url, params=params)

            if response.status_code == 200:
                data = response.json()

                if 'result' in data and 'records' in data['result']:
                    records = data['result']['records']

                    if len(records) == 0:
                        break  # No hay m√°s registros

                    all_records.extend(records)
                    print(f"   Descargados {len(all_records)} registros...")

                    offset += limit  # Siguiente p√°gina
                else:
                    break
            else:
                print(f"‚ùå Error: {response.status_code}")
                break

        except Exception as e:
            print(f"‚ùå Error: {e}")
            break

    if all_records:
        df = pd.DataFrame(all_records)
        
        # Guardar copia local
       
        print(f"\n‚úÖ Total descargado: {len(df)} registros")
        
        return df
    else:
        print("‚ùå No se pudieron descargar datos")
        return None

print("Funci√≥n de descarga definida")
```

```python
def analisis_datos_faltantes_psdata(df):
    """
    An√°lisis exhaustivo de datos faltantes para PSData
    """
    print(f"\nüìä AN√ÅLISIS DETALLADO DE DATOS FALTANTES")
    print("=" * 50)
    
    # Conteo de nulos por columna
    nulos_por_columna = df.isnull().sum()
    porcentaje_nulos = (nulos_por_columna / len(df) * 100).round(2)
    
    # Crear DataFrame de resumen
    resumen_nulos = pd.DataFrame({
        'Columna': nulos_por_columna.index,
        'Valores_Nulos': nulos_por_columna.values,
        'Porcentaje_Nulos': porcentaje_nulos.values,
        'Valores_Completos': len(df) - nulos_por_columna.values
    }).sort_values('Porcentaje_Nulos', ascending=False)
    
    print("üìã Resumen de datos faltantes:")
    display(resumen_nulos)
    
    # Visualizaci√≥n de datos faltantes
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Gr√°fico de barras de porcentaje de nulos
    columnas_con_nulos = resumen_nulos[resumen_nulos['Porcentaje_Nulos'] > 0]
    if len(columnas_con_nulos) > 0:
        top_nulos = columnas_con_nulos.head(10)  # Top 10 con m√°s nulos
        ax1.barh(top_nulos['Columna'], top_nulos['Porcentaje_Nulos'])
        ax1.set_title('Top 10 Columnas con M√°s Datos Faltantes')
        ax1.set_xlabel('Porcentaje (%)')
    else:
        ax1.text(0.5, 0.5, 'No hay datos faltantes', ha='center', va='center', transform=ax1.transAxes)
        ax1.set_title('Sin Datos Faltantes')
    
    # Heatmap de patrones de nulos (muestra)
    if df.isnull().sum().sum() > 0:
        # Tomar muestra para visualizaci√≥n si el dataset es muy grande
        muestra_size = min(1000, len(df))
        df_muestra = df.sample(n=muestra_size, random_state=42)
        sns.heatmap(df_muestra.isnull(), cbar=True, yticklabels=False, cmap='viridis', ax=ax2)
        ax2.set_title(f'Patr√≥n de Datos Faltantes (muestra de {muestra_size})')
    else:
        ax2.text(0.5, 0.5, 'Dataset Completo', ha='center', va='center', transform=ax2.transAxes)
        ax2.set_title('Sin Datos Faltantes')
    
    plt.tight_layout()
    plt.show()
    
    return resumen_nulos

# Ejecutar an√°lisis de datos faltantes
if df_psdata_raw is not None:
    resumen_nulos_psdata = analisis_datos_faltantes_psdata(df_psdata_raw)
```

```python
def limpiar_datos_cobertura(df):
    """
    Limpia y procesa los datos de cobertura m√≥vil
    """
    print(f"\nüßπ LIMPIEZA Y PROCESAMIENTO DE DATOS")
    print("-" * 45)
    
    df_clean = df.copy()
    
    # 1. Identificar columnas de cobertura autom√°ticamente
    columnas_cobertura = []
    operadores = ['claro', 'movistar', 'tigo', 'wom']
    
    print("üîç Buscando columnas de cobertura...")
    for col in df.columns:
        col_lower = col.lower()
        if any(op in col_lower for op in operadores) and ('cob' in col_lower or 'area' in col_lower):
            columnas_cobertura.append(col)
    
    print(f"üì° Columnas de cobertura identificadas:")
    for col in columnas_cobertura:
        print(f"   ‚Ä¢ {col}")
    
    # 2. Limpiar y convertir columnas de cobertura a num√©rico
    print(f"\nüîß Convirtiendo columnas de cobertura a valores num√©ricos...")
    
    for col in columnas_cobertura:
        print(f"   Procesando {col}...")
        
        # Convertir a string primero
        df_clean[col] = df_clean[col].astype(str)
        
        # Limpiar valores
        df_clean[col] = df_clean[col].str.replace(',', '.')  # Comas por puntos
        df_clean[col] = df_clean[col].str.replace(' ', '')   # Eliminar espacios
        df_clean[col] = df_clean[col].replace(['nan', 'NaN', 'null', 'NULL', '', 'None'], np.nan)
        
        # Convertir a num√©rico
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
        
        # Estad√≠sticas de conversi√≥n
        valores_validos = df_clean[col].notna().sum()
        total = len(df_clean)
        porcentaje = (valores_validos / total) * 100
        
        print(f"     {valores_validos}/{total} valores v√°lidos ({porcentaje:.1f}%)")
    
    # 3. Limpiar variables geogr√°ficas
    print(f"\nüó∫Ô∏è Procesando variables geogr√°ficas...")
    
    # Municipio
    if 'municipio' in df_clean.columns:
        df_clean['municipio_limpio'] = (df_clean['municipio']
                                       .astype(str)
                                       .str.upper()
                                       .str.strip()
                                       .str.replace('√Å', 'A')
                                       .str.replace('√â', 'E')
                                       .str.replace('√ç', 'I')
                                       .str.replace('√ì', 'O')
                                       .str.replace('√ö', 'U')
                                       .str.replace('√ë', 'N'))
        print("   ‚úÖ Municipio procesado")
    
    # Departamento
    if 'departamento' in df_clean.columns:
        df_clean['departamento_limpio'] = (df_clean['departamento']
                                          .astype(str)
                                          .str.upper()
                                          .str.strip())
        print("   ‚úÖ Departamento procesado")
    
    # 4. Crear m√©tricas derivadas
    print(f"\nüìä Creando m√©tricas derivadas...")
    
    if len(columnas_cobertura) > 0:
        # Cobertura promedio por registro
        df_clean['cobertura_promedio'] = df_clean[columnas_cobertura].mean(axis=1)
        
        # Cobertura total
        df_clean['cobertura_total'] = df_clean[columnas_cobertura].sum(axis=1)
        
        # N√∫mero de operadores con datos v√°lidos
        df_clean['operadores_activos'] = df_clean[columnas_cobertura].notna().sum(axis=1)
        
        print(f"   ‚úÖ M√©tricas derivadas creadas:")
        print(f"      ‚Ä¢ Cobertura promedio")
        print(f"      ‚Ä¢ Cobertura total")
        print(f"      ‚Ä¢ Operadores activos")
    
    print(f"\n‚úÖ Datos procesados exitosamente")
    print(f"   Registros finales: {len(df_clean)}")
    
    return df_clean, columnas_cobertura

# Aplicar limpieza
if df_psdata_raw is not None:
    df_psdata_clean, columnas_cobertura = limpiar_datos_cobertura(df_psdata_raw)
    
    print(f"\nüìã Muestra de datos limpios:")
    display(df_psdata_clean.head())
```

```python
def detectar_outliers_cobertura(df, columnas_cobertura):
    """
    Detecta outliers en las variables de cobertura
    """
    print(f"\nüìê DETECCI√ìN DE OUTLIERS - VARIABLES DE COBERTURA")
    print("-" * 55)
    
    if len(columnas_cobertura) == 0:
        print("‚ùå No hay columnas de cobertura para analizar")
        return None
    
    outliers_info = {}
    
    for col in columnas_cobertura:
        if col in df.columns:
            valores = df[col].dropna()
            
            if len(valores) == 0:
                print(f"‚ùå {col}: No hay valores v√°lidos")
                continue
            
            print(f"\nüìä AN√ÅLISIS DE: {col.upper()}")
            print("-" * 30)
            
            # M√©todo IQR
            Q1 = valores.quantile(0.25)
            Q3 = valores.quantile(0.75)
            IQR = Q3 - Q1
            limite_inferior = Q1 - 1.5 * IQR
            limite_superior = Q3 + 1.5 * IQR
            
            outliers_iqr = valores[(valores < limite_inferior) | (valores > limite_superior)]
            
            # M√©todo Z-score
            if valores.std() > 0:
                z_scores = np.abs((valores - valores.mean()) / valores.std())
                outliers_zscore = valores[z_scores > 3]
            else:
                outliers_zscore = pd.Series(dtype=float)
            
            print(f"Estad√≠sticas b√°sicas:")
            print(f"   Media: {valores.mean():.4f}")
            print(f"   Mediana: {valores.median():.4f}")
            print(f"   Desv. est√°ndar: {valores.std():.4f}")
            print(f"   Rango: {valores.min():.4f} - {valores.max():.4f}")
            
            print(f"\nOutliers detectados:")
            print(f"   M√©todo IQR: {len(outliers_iqr)} outliers")
            print(f"   M√©todo Z-score: {len(outliers_zscore)} outliers")
            
            outliers_info[col] = {
                'outliers_iqr': outliers_iqr,
                'outliers_zscore': outliers_zscore,
                'limites_iqr': (limite_inferior, limite_superior)
            }
    
    # Visualizaci√≥n de outliers
    if len(columnas_cobertura) > 0:
        n_cols = min(len(columnas_cobertura), 4)  # M√°ximo 4 columnas
        n_rows = 2
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 10))
        
        if n_cols == 1:
            axes = axes.reshape(-1, 1)
        elif len(columnas_cobertura) == 1:
            axes = [[axes[0]], [axes[1]]]
        
        for i, col in enumerate(columnas_cobertura[:n_cols]):
            if col in df.columns:
                valores = df[col].dropna()
                
                if len(valores) > 0:
                    # Histograma
                    axes[0][i].hist(valores, bins=30, edgecolor='black', alpha=0.7)
                    axes[0][i].axvline(valores.mean(), color='red', linestyle='--', label=f'Media')
                    axes[0][i].axvline(valores.median(), color='green', linestyle='--', label=f'Mediana')
                    axes[0][i].set_title(f'Distribuci√≥n - {col}')
                    axes[0][i].legend()
                    
                    # Boxplot
                    axes[1][i].boxplot(valores)
                    axes[1][i].set_title(f'Boxplot - {col}')
        
        plt.tight_layout()
        plt.show()
    
    return outliers_info

# Ejecutar detecci√≥n de outliers
if 'df_psdata_clean' in locals() and 'columnas_cobertura' in locals():
    outliers_info_psdata = detectar_outliers_cobertura(df_psdata_clean, columnas_cobertura)
```

```python
def analisis_exploratorio_cobertura(df, columnas_cobertura):
    """
    An√°lisis exploratorio de datos de cobertura m√≥vil
    """
    print(f"\nüìà AN√ÅLISIS EXPLORATORIO AVANZADO - COBERTURA M√ìVIL")
    print("=" * 60)
    
    if len(columnas_cobertura) == 0:
        print("‚ùå No hay columnas de cobertura para analizar")
        return None
    
    # 1. Estad√≠sticas descriptivas completas
    print(f"\nüìä ESTAD√çSTICAS DESCRIPTIVAS - VARIABLES DE COBERTURA:")
    estadisticas_cobertura = df[columnas_cobertura].describe()
    display(estadisticas_cobertura)
    
    # 2. An√°lisis por municipios
    cobertura_municipios = None
    if 'municipio_limpio' in df.columns:
        print(f"\nüèõÔ∏è AN√ÅLISIS POR MUNICIPIOS:")
        municipios_unicos = df['municipio_limpio'].nunique()
        print(f"Total de municipios √∫nicos: {municipios_unicos}")
        
        # Cobertura promedio por municipio
        cobertura_municipios = df.groupby('municipio_limpio')[columnas_cobertura].mean().reset_index()
        cobertura_municipios['cobertura_promedio'] = cobertura_municipios[columnas_cobertura].mean(axis=1)
        
        print(f"\nTop 10 municipios con mejor cobertura promedio:")
        top_municipios = cobertura_municipios.nlargest(10, 'cobertura_promedio')
        display(top_municipios[['municipio_limpio', 'cobertura_promedio'] + columnas_cobertura])
        
        print(f"\nTop 10 municipios con menor cobertura promedio:")
        bottom_municipios = cobertura_municipios.nsmallest(10, 'cobertura_promedio')
        display(bottom_municipios[['municipio_limpio', 'cobertura_promedio'] + columnas_cobertura])
    
    # 3. An√°lisis por departamentos
    if 'departamento_limpio' in df.columns:
        print(f"\nüó∫Ô∏è AN√ÅLISIS POR DEPARTAMENTOS:")
        departamentos_unicos = df['departamento_limpio'].nunique()
        print(f"Total de departamentos √∫nicos: {departamentos_unicos}")
        
        cobertura_deptos = df.groupby('departamento_limpio')[columnas_cobertura].mean().reset_index()
        cobertura_deptos['cobertura_promedio'] = cobertura_deptos[columnas_cobertura].mean(axis=1)
        
        print(f"\nCobertura promedio por departamento:")
        cobertura_deptos_sorted = cobertura_deptos.sort_values('cobertura_promedio', ascending=False)
        display(cobertura_deptos_sorted)
    
    # 4. Visualizaciones comprehensivas
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    
    # Distribuci√≥n de cobertura por operador
    df[columnas_cobertura].boxplot(ax=axes[0,0])
    axes[0,0].set_title('Distribuci√≥n de Cobertura por Operador')
    axes[0,0].tick_params(axis='x', rotation=45)
    
    # Correlaci√≥n entre operadores
    if len(columnas_cobertura) > 1:
        correlacion = df[columnas_cobertura].corr()
        sns.heatmap(correlacion, annot=True, cmap='coolwarm', ax=axes[0,1])
        axes[0,1].set_title('Correlaci√≥n entre Operadores')
    
    # Histograma de cobertura promedio
    if 'cobertura_promedio' in df.columns:
        valores_promedio = df['cobertura_promedio'].dropna()
        axes[0,2].hist(valores_promedio, bins=30, edgecolor='black')
        axes[0,2].set_title('Distribuci√≥n de Cobertura Promedio')
        axes[0,2].set_xlabel('Cobertura Promedio')
    
    # An√°lisis por departamento (si disponible)
    if 'departamento_limpio' in df.columns and len(columnas_cobertura) > 0:
        cobertura_dept = df.groupby('departamento_limpio')[columnas_cobertura].mean()
        if len(cobertura_dept) > 0:
            cobertura_dept.mean(axis=1).sort_values(ascending=True).tail(15).plot(kind='barh', ax=axes[1,0])
            axes[1,0].set_title('Top 15 Departamentos - Cobertura Promedio')
            axes[1,0].set_xlabel('Cobertura Promedio')
    
    # Scatter plot de operadores
    if len(columnas_cobertura) >= 2:
        col1, col2 = columnas_cobertura[0], columnas_cobertura[1]
        data1 = df[col1].dropna()
        data2 = df[col2].dropna()
        if len(data1) > 0 and len(data2) > 0:
            axes[1,1].scatter(df[col1], df[col2], alpha=0.5)
            axes[1,1].set_xlabel(col1)
            axes[1,1].set_ylabel(col2)
            axes[1,1].set_title(f'Correlaci√≥n {col1} vs {col2}')
    
    # N√∫mero de operadores activos por municipio
    if 'operadores_activos' in df.columns:
        operadores_count = df['operadores_activos'].value_counts().sort_index()
        if len(operadores_count) > 0:
            axes[1,2].bar(operadores_count.index, operadores_count.values)
            axes[1,2].set_title('N√∫mero de Operadores Activos')
            axes[1,2].set_xlabel('N√∫mero de Operadores')
            axes[1,2].set_ylabel('Cantidad de Registros')
    
    plt.tight_layout()
    plt.show()
    
    return cobertura_municipios

# Ejecutar an√°lisis exploratorio
if 'df_psdata_clean' in locals() and 'columnas_cobertura' in locals():
    resultado_exploratorio = analisis_exploratorio_cobertura(df_psdata_clean, columnas_cobertura)
```

```python
def analisis_estadistico_profundo_psdata(df, columnas_cobertura):
    """
    An√°lisis estad√≠stico profundo de las variables de cobertura
    """
    print(f"\nüî¨ AN√ÅLISIS ESTAD√çSTICO PROFUNDO - PSDATA")
    print("=" * 45)
    
    if len(columnas_cobertura) == 0:
        print("‚ùå No hay columnas de cobertura para analizar")
        return
    
    # An√°lisis de normalidad para cada variable de cobertura
    for col in columnas_cobertura:
        if col in df.columns:
            valores = df[col].dropna()
            
            if len(valores) > 0:
                print(f"\nüìê AN√ÅLISIS DE: {col.upper()}")
                print("-" * 30)
                
                # Estad√≠sticas descriptivas
                print(f"Estad√≠sticas descriptivas:")
                print(f"  N: {len(valores)}")
                print(f"  Media: {valores.mean():.6f}")
                print(f"  Mediana: {valores.median():.6f}")
                print(f"  Desv. est√°ndar: {valores.std():.6f}")
                print(f"  Varianza: {valores.var():.6f}")
                print(f"  Asimetr√≠a (skewness): {valores.skew():.6f}")
                print(f"  Curtosis: {valores.kurtosis():.6f}")
                
                # Interpretaci√≥n de asimetr√≠a
                skew_val = valores.skew()
                if abs(skew_val) < 0.5:
                    skew_interp = "aproximadamente sim√©trica"
                elif skew_val > 0.5:
                    skew_interp = "asim√©trica positiva (cola derecha)"
                else:
                    skew_interp = "asim√©trica negativa (cola izquierda)"
                print(f"  Interpretaci√≥n: Distribuci√≥n {skew_interp}")
                
                # Prueba de normalidad
                if len(valores) >= 8:
                    if len(valores) <= 5000:
                        stat, p_value = stats.shapiro(valores)
                        test_name = "Shapiro-Wilk"
                    else:
                        stat, p_value = stats.normaltest(valores)
                        test_name = "D'Agostino"
                    
                    print(f"\nüß™ Prueba de normalidad ({test_name}):")
                    print(f"  Estad√≠stico: {stat:.6f}")
                    print(f"  p-valor: {p_value:.6f}")
                    
                    if p_value > 0.05:
                        print(f"  ‚úÖ Los datos parecen seguir una distribuci√≥n normal (p > 0.05)")
                    else:
                        print(f"  ‚ùå Los datos NO siguen una distribuci√≥n normal (p ‚â§ 0.05)")

# Ejecutar an√°lisis estad√≠stico profundo
if 'df_psdata_clean' in locals() and 'columnas_cobertura' in locals():
    analisis_estadistico_profundo_psdata(df_psdata_clean, columnas_cobertura)
```

```python
def preparar_para_integracion_dane(df):
    """
    Prepara los datos de PSData para integraci√≥n con datos del DANE
    """
    print(f"\nüîó PREPARACI√ìN PARA INTEGRACI√ìN CON DANE")
    print("-" * 50)
    
    # Identificar columnas de cobertura
    columnas_a_agregar = []
    
    # Variables de cobertura
    for col in df.columns:
        if any(op in col.lower() for op in ['claro', 'movistar', 'tigo', 'wom']) and ('cob' in col.lower() or 'area' in col.lower()):
            columnas_a_agregar.append(col)
    
    # Variables derivadas
    vars_derivadas = ['cobertura_promedio', 'cobertura_total', 'operadores_activos']
    for var in vars_derivadas:
        if var in df.columns:
            columnas_a_agregar.append(var)
    
    if len(columnas_a_agregar) == 0:
        print("‚ùå No se encontraron variables de cobertura para agregar")
        return None
    
    print(f"üìã Variables a agregar: {columnas_a_agregar}")
    
    # Agregar por municipio usando promedio
    df_agregado = df.groupby(columna_municipio)[columnas_a_agregar].mean().reset_index()
    
    # Renombrar columna de municipio para consistencia
    df_agregado = df_agregado.rename(columns={columna_municipio: 'municipio'})
    
    # Limpiar nombres de municipios para matching
    df_agregado['municipio_para_merge'] = (df_agregado['municipio']
                                          .str.upper()
                                          .str.strip()
                                          .str.replace('√Å', 'A')
                                          .str.replace('√â', 'E')
                                          .str.replace('√ç', 'I')
                                          .str.replace('√ì', 'O')
                                          .str.replace('√ö', 'U')
                                          .str.replace('√ë', 'N'))
    
    # Estad√≠sticas del agregado
    print(f"\nüìà Resultados de la agregaci√≥n:")
    print(f"   Municipios √∫nicos: {len(df_agregado)}")
    print(f"   Variables de cobertura: {len(columnas_a_agregar)}")
    
    # Verificar completitud
    print(f"\nüîç Verificaci√≥n de completitud:")
    for col in columnas_a_agregar:
        valores_validos = df_agregado[col].notna().sum()
        porcentaje = (valores_validos / len(df_agregado)) * 100
        print(f"   {col}: {valores_validos}/{len(df_agregado)} ({porcentaje:.1f}%)")
    
    return df_agregado

# Preparar datos para integraci√≥n
if 'df_psdata_clean' in locals():
    df_psdata_para_merge = preparar_para_integracion_dane(df_psdata_clean)
    
    if df_psdata_para_merge is not None:
        print(f"\nüìã Muestra de datos preparados para merge:")
        display(df_psdata_para_merge.head(10))
```

```python
def preparar_para_integracion_dane_corregida(df):
    """
    Prepara los datos de PSData para integraci√≥n con datos del DANE - VERSI√ìN CORREGIDA
    """
    print(f"\nüîó PREPARACI√ìN PARA INTEGRACI√ìN CON DANE")
    print("-" * 50)
    
    # Identificar columna de municipio
    columna_municipio = None
    posibles_municipio = ['municipio', 'municipio_limpio', 'MUNICIPIO']
    for col in posibles_municipio:
        if col in df.columns:
            columna_municipio = col
            break
    
    if columna_municipio is None:
        print("‚ùå No se encontr√≥ columna de municipio")
        print("üìã Columnas disponibles:")
        for col in df.columns:
            print(f"   ‚Ä¢ {col}")
        return None
    
    print(f"‚úÖ Usando columna de municipio: {columna_municipio}")
    
    # Identificar columnas de cobertura
    columnas_a_agregar = []
    
    # Variables de cobertura
    for col in df.columns:
        if any(op in col.lower() for op in ['claro', 'movistar', 'tigo', 'wom']) and ('cob' in col.lower() or 'area' in col.lower()):
            columnas_a_agregar.append(col)
    
    # Variables derivadas
    vars_derivadas = ['cobertura_promedio', 'cobertura_total', 'operadores_activos']
    for var in vars_derivadas:
        if var in df.columns:
            columnas_a_agregar.append(var)
    
    if len(columnas_a_agregar) == 0:
        print("‚ùå No se encontraron variables de cobertura para agregar")
        return None
    
    print(f"üìã Variables a agregar: {columnas_a_agregar}")
    
    # Agregar por municipio usando promedio
    df_agregado = df.groupby(columna_municipio)[columnas_a_agregar].mean().reset_index()
    
    # Renombrar columna de municipio para consistencia
    df_agregado = df_agregado.rename(columns={columna_municipio: 'municipio'})
    
    # Limpiar nombres de municipios para matching
    df_agregado['municipio_para_merge'] = (df_agregado['municipio']
                                          .astype(str)
                                          .str.upper()
                                          .str.strip()
                                          .str.replace('√Å', 'A')
                                          .str.replace('√â', 'E')
                                          .str.replace('√ç', 'I')
                                          .str.replace('√ì', 'O')
                                          .str.replace('√ö', 'U')
                                          .str.replace('√ë', 'N'))
    
    # Estad√≠sticas del agregado
    print(f"\nüìà Resultados de la agregaci√≥n:")
    print(f"   Municipios √∫nicos: {len(df_agregado)}")
    print(f"   Variables de cobertura: {len(columnas_a_agregar)}")
    
    # Verificar completitud
    print(f"\nüîç Verificaci√≥n de completitud:")
    for col in columnas_a_agregar:
        valores_validos = df_agregado[col].notna().sum()
        porcentaje = (valores_validos / len(df_agregado)) * 100
        print(f"   {col}: {valores_validos}/{len(df_agregado)} ({porcentaje:.1f}%)")
    
    return df_agregado

df_psdata_para_merge = preparar_para_integracion_dane_corregida(df_psdata_clean)
```

```python
def analisis_comparativo_operadores(df, columnas_cobertura):
    """
    An√°lisis comparativo detallado entre operadores
    """
    print(f"\nüìä AN√ÅLISIS COMPARATIVO DE OPERADORES")
    print("=" * 45)
    
    if len(columnas_cobertura) == 0:
        print("‚ùå No hay columnas de cobertura para analizar")
        return None
    
    # 1. Estad√≠sticas por operador
    print(f"\nüìà Estad√≠sticas por operador:")
    estadisticas_operadores = df[columnas_cobertura].describe()
    display(estadisticas_operadores)
    
    # 2. Rankings de operadores
    print(f"\nüèÜ Rankings de operadores:")
    
    # Por cobertura promedio
    cobertura_promedio = df[columnas_cobertura].mean().sort_values(ascending=False)
    print(f"\nPor cobertura promedio:")
    for i, (operador, valor) in enumerate(cobertura_promedio.items(), 1):
        print(f"   {i}. {operador}: {valor:.4f}")
    
    # Por consistencia (menor desviaci√≥n est√°ndar)
    consistencia = df[columnas_cobertura].std().sort_values(ascending=True)
    print(f"\nPor consistencia (menor variabilidad):")
    for i, (operador, valor) in enumerate(consistencia.items(), 1):
        print(f"   {i}. {operador}: œÉ = {valor:.4f}")
    
    # 3. An√°lisis de mercado
    print(f"\nüìä An√°lisis de participaci√≥n de mercado:")
    
    # Calcular participaci√≥n de mercado
    total_cobertura = df[columnas_cobertura].sum().sum()
    if total_cobertura > 0:
        participacion = df[columnas_cobertura].sum() / total_cobertura * 100
        
        for operador, porcentaje in participacion.sort_values(ascending=False).items():
            print(f"   {operador}: {porcentaje:.2f}%")
    else:
        participacion = None
        print("   No se pudo calcular participaci√≥n de mercado")
    
    # 4. Visualizaci√≥n comparativa
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # Comparaci√≥n de medias
    cobertura_promedio.plot(kind='bar', ax=axes[0,0])
    axes[0,0].set_title('Cobertura Promedio por Operador')
    axes[0,0].set_ylabel('Cobertura Promedio')
    axes[0,0].tick_params(axis='x', rotation=45)
    
    # Comparaci√≥n de variabilidad
    consistencia.plot(kind='bar', ax=axes[0,1], color='orange')
    axes[0,1].set_title('Variabilidad por Operador (Desv. Est√°ndar)')
    axes[0,1].set_ylabel('Desviaci√≥n Est√°ndar')
    axes[0,1].tick_params(axis='x', rotation=45)
    
    # Distribuciones superpuestas
    for col in columnas_cobertura:
        valores = df[col].dropna()
        if len(valores) > 0:
            axes[1,0].hist(valores, alpha=0.5, label=col, bins=30)
    axes[1,0].set_title('Distribuciones de Cobertura Superpuestas')
    axes[1,0].legend()
    axes[1,0].set_xlabel('Cobertura')
    
    # Participaci√≥n de mercado
    if participacion is not None and len(participacion) > 0:
        axes[1,1].pie(participacion.values, labels=participacion.index, autopct='%1.1f%%')
        axes[1,1].set_title('Participaci√≥n de Mercado')
    
    plt.tight_layout()
    plt.show()
    
    return {
        'cobertura_promedio': cobertura_promedio,
        'consistencia': consistencia,
        'participacion_mercado': participacion
    }

# Ejecutar an√°lisis comparativo
if 'df_psdata_clean' in locals() and 'columnas_cobertura' in locals():
    analisis_operadores = analisis_comparativo_operadores(df_psdata_clean, columnas_cobertura)
```

```python
def validar_calidad_psdata(df_original, df_procesado, df_agregado):
    """
    Validaci√≥n de calidad de los datos procesados
    """
    print(f"\n‚úÖ VALIDACI√ìN DE CALIDAD - PSDATA")
    print("-" * 40)
    
    # 1. Comparaci√≥n de registros
    print(f"üìä Comparaci√≥n de registros:")
    print(f"   Datos originales: {len(df_original)} registros")
    print(f"   Datos procesados: {len(df_procesado)} registros")
    if df_agregado is not None:
        print(f"   Datos agregados por municipio: {len(df_agregado)} registros")
    
    # 2. Verificar consistencia de variables de cobertura
    columnas_cobertura = [col for col in df_procesado.columns 
                         if any(op in col.lower() for op in ['claro', 'movistar', 'tigo', 'wom']) 
                         and ('cob' in col.lower() or 'area' in col.lower())]
    
    print(f"\nüì° Validaci√≥n de variables de cobertura:")
    for col in columnas_cobertura:
        if col in df_procesado.columns:
            valores = df_procesado[col].dropna()
            if len(valores) > 0:
                min_val = valores.min()
                max_val = valores.max()
                valores_negativos = (valores < 0).sum()
                valores_extremos = (valores > 100).sum()
                
                print(f"   {col}:")
                print(f"     Rango: {min_val:.4f} - {max_val:.4f}")
                print(f"     Valores negativos: {valores_negativos}")
                print(f"     Valores > 100: {valores_extremos}")
    
    # 3. Verificar duplicados
    if df_agregado is not None and 'municipio' in df_agregado.columns:
        duplicados = df_agregado['municipio'].duplicated().sum()
        print(f"\nüîç Verificaci√≥n de duplicados:")
        print(f"   Municipios duplicados en datos agregados: {duplicados}")
    
    # 4. Verificar cobertura de municipios principales
    if df_agregado is not None and 'municipio' in df_agregado.columns:
        municipios_con_datos = df_agregado['municipio'].notna().sum()
        print(f"\nüó∫Ô∏è Cobertura territorial:")
        print(f"   Municipios con datos v√°lidos: {municipios_con_datos}")
        
        # Verificar principales municipios
        principales_municipios = ['BOGOT√Å', 'MEDELL√çN', 'CALI', 'BARRANQUILLA', 'CARTAGENA']
        municipios_disponibles = df_agregado['municipio'].str.upper().values
        
        for municipio in principales_municipios:
            tiene_datos = any(municipio in mun for mun in municipios_disponibles)
            status = "‚úÖ" if tiene_datos else "‚ùå"
            print(f"   {municipio}: {status}")
    
    return True

# Ejecutar validaci√≥n
if all(var in locals() for var in ['df_psdata_raw', 'df_psdata_clean']):
    if 'df_psdata_para_merge' in locals() and df_psdata_para_merge is not None:
        validacion_ok = validar_calidad_psdata(df_psdata_raw, df_psdata_clean, df_psdata_para_merge)
    else:
        validacion_ok = validar_calidad_psdata(df_psdata_raw, df_psdata_clean, None)

validacion_ok = validar_calidad_psdata(df_psdata_raw, df_psdata_clean, df_psdata_para_merge)
```
