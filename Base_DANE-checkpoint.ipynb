```python
def procesar_datos_municipios(df_raw):
    """
    Procesa y limpia los datos de municipios del DANE
    """
    df = df_raw.copy()
    
    print(f"\nüßπ PROCESANDO Y LIMPIANDO DATOS")
    print("-" * 40)
    
    # Identificar columnas importantes del DANE
    columnas_identificadas = {}
    
    # Buscar c√≥digo DIVIPOLA
    posibles_codigo = ['MPIO_CCNCT', 'MPIOS', 'COD_MPIO', 'CODIGO']
    for col in posibles_codigo:
        if col in df.columns:
            columnas_identificadas['codigo'] = col
            break
    
    # Buscar nombre municipio
    posibles_nombre = ['MPIO_CNMBR', 'NOMBRE_MPI', 'MUNICIPIO', 'NOMBRE']
    for col in posibles_nombre:
        if col in df.columns:
            columnas_identificadas['municipio'] = col
            break
    
    # Buscar departamento
    posibles_depto = ['DPTO_CNMBR', 'DEPARTAMEN', 'DEPTO', 'DPTO_CCDGO']
    for col in posibles_depto:
        if col in df.columns:
            columnas_identificadas['departamento'] = col
            break
    
    # Buscar √°rea
    posibles_area = ['MPIO_CAREA', 'AREA', 'AREA_KM2']
    for col in posibles_area:
        if col in df.columns:
            columnas_identificadas['area'] = col
            break
    
    print(f"üìù Columnas identificadas:")
    for tipo, col in columnas_identificadas.items():
        print(f"   {tipo.capitalize()}: {col}")
    
    # Crear columnas estandarizadas
    if 'codigo' in columnas_identificadas:
        df['codigo_municipio'] = df[columnas_identificadas['codigo']].astype(str).str.zfill(5)
    
    if 'municipio' in columnas_identificadas:
        df['municipio_limpio'] = (df[columnas_identificadas['municipio']]
                                 .str.upper()
                                 .str.strip()
                                 .str.replace('√Å', 'A')
                                 .str.replace('√â', 'E')
                                 .str.replace('√ç', 'I')
                                 .str.replace('√ì', 'O')
                                 .str.replace('√ö', 'U')
                                 .str.replace('√ë', 'N'))
    
    if 'departamento' in columnas_identificadas:
        df['departamento_limpio'] = (df[columnas_identificadas['departamento']]
                                   .str.upper()
                                   .str.strip())
    
    # Convertir √°rea a num√©rico si existe
    if 'area' in columnas_identificadas:
        df['area_km2'] = pd.to_numeric(df[columnas_identificadas['area']], errors='coerce')
    
    # Validar c√≥digos DIVIPOLA
    if 'codigo_municipio' in df.columns:
        # Filtrar c√≥digos v√°lidos (5 d√≠gitos)
        mask_validos = df['codigo_municipio'].str.match(r'^\d{5}$', na=False)
        df_limpio = df[mask_validos].copy()
        
        removidos = len(df) - len(df_limpio)
        if removidos > 0:
            print(f"‚ö†Ô∏è Removidos {removidos} registros con c√≥digos DIVIPOLA inv√°lidos")
        
        print(f"‚úÖ Registros v√°lidos: {len(df_limpio)}")
        return df_limpio
    
    return df

print("üõ†Ô∏è Funciones de procesamiento definidas")
```

```python
def analisis_datos_faltantes(df):
    """
    An√°lisis exhaustivo de datos faltantes
    """
    print(f"\nüìä AN√ÅLISIS DETALLADO DE DATOS FALTANTES")
    print("=" * 50)
    
    # Conteo de nulos por columna
    nulos_por_columna = df.isnull().sum()
    porcentaje_nulos = (nulos_por_columna / len(df) * 100).round(2)
    
    # Crear DataFrame de resumen
    resumen_nulos = pd.DataFrame({
        'Columna': nulos_por_columna.index,
        'Valores_Nulos': nulos_por_columna.values,
        'Porcentaje_Nulos': porcentaje_nulos.values,
        'Valores_Completos': len(df) - nulos_por_columna.values
    }).sort_values('Porcentaje_Nulos', ascending=False)
    
    print("üìã Resumen de datos faltantes:")
    display(resumen_nulos)
    
    # Visualizaci√≥n de datos faltantes
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    
    # Gr√°fico de barras de porcentaje de nulos
    columnas_con_nulos = resumen_nulos[resumen_nulos['Porcentaje_Nulos'] > 0]
    if len(columnas_con_nulos) > 0:
        ax1.bar(columnas_con_nulos['Columna'], columnas_con_nulos['Porcentaje_Nulos'])
        ax1.set_title('Porcentaje de Datos Faltantes por Columna')
        ax1.tick_params(axis='x', rotation=45)
        ax1.set_ylabel('Porcentaje (%)')
    else:
        ax1.text(0.5, 0.5, 'No hay datos faltantes', ha='center', va='center', transform=ax1.transAxes)
        ax1.set_title('Sin Datos Faltantes')
    
    # Heatmap de patrones de nulos
    if df.isnull().sum().sum() > 0:
        sns.heatmap(df.isnull(), cbar=True, yticklabels=False, cmap='viridis', ax=ax2)
        ax2.set_title('Patr√≥n de Datos Faltantes')
    else:
        ax2.text(0.5, 0.5, 'Dataset Completo', ha='center', va='center', transform=ax2.transAxes)
        ax2.set_title('Sin Datos Faltantes')
    
    plt.tight_layout()
    plt.show()
    
    # Estrategias de manejo
    print(f"\nüõ†Ô∏è ESTRATEGIAS DE MANEJO:")
    for _, row in columnas_con_nulos.iterrows():
        if row['Porcentaje_Nulos'] > 50:
            print(f"‚Ä¢ {row['Columna']}: Considerar eliminaci√≥n (>{row['Porcentaje_Nulos']:.1f}% nulos)")
        elif row['Porcentaje_Nulos'] > 20:
            print(f"‚Ä¢ {row['Columna']}: Imputaci√≥n requerida ({row['Porcentaje_Nulos']:.1f}% nulos)")
        else:
            print(f"‚Ä¢ {row['Columna']}: Nulos manejables ({row['Porcentaje_Nulos']:.1f}% nulos)")
    
    return resumen_nulos

# Ejecutar an√°lisis de datos faltantes
if df_municipios is not None:
    resumen_nulos = analisis_datos_faltantes(df_municipios)
```

```python
def detectar_outliers(df, columna_numerica):
    """
    Detecta outliers usando m√©todo IQR y Z-score
    """
    if columna_numerica not in df.columns:
        print(f"‚ùå Columna {columna_numerica} no encontrada")
        return None
    
    print(f"\nüìê DETECCI√ìN DE OUTLIERS - {columna_numerica.upper()}")
    print("-" * 45)
    
    # Filtrar valores no nulos
    valores = df[columna_numerica].dropna()
    
    if len(valores) == 0:
        print("‚ùå No hay valores num√©ricos v√°lidos para analizar")
        return None
    
    # M√©todo IQR
    Q1 = valores.quantile(0.25)
    Q3 = valores.quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    outliers_iqr = valores[(valores < limite_inferior) | (valores > limite_superior)]
    
    # M√©todo Z-score
    z_scores = np.abs((valores - valores.mean()) / valores.std())
    outliers_zscore = valores[z_scores > 3]
    
    print(f"üìä Estad√≠sticas b√°sicas:")
    print(f"   Media: {valores.mean():.2f}")
    print(f"   Mediana: {valores.median():.2f}")
    print(f"   Desviaci√≥n est√°ndar: {valores.std():.2f}")
    print(f"   Rango: {valores.min():.2f} - {valores.max():.2f}")
    
    print(f"\nüéØ Detecci√≥n de outliers:")
    print(f"   M√©todo IQR (Q1-1.5*IQR, Q3+1.5*IQR): {len(outliers_iqr)} outliers")
    print(f"   L√≠mites IQR: [{limite_inferior:.2f}, {limite_superior:.2f}]")
    print(f"   M√©todo Z-score (|z| > 3): {len(outliers_zscore)} outliers")
    
    # Visualizaci√≥n
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
    
    # Histograma
    ax1.hist(valores, bins=30, edgecolor='black', alpha=0.7)
    ax1.axvline(valores.mean(), color='red', linestyle='--', label=f'Media: {valores.mean():.2f}')
    ax1.axvline(valores.median(), color='green', linestyle='--', label=f'Mediana: {valores.median():.2f}')
    ax1.set_title(f'Distribuci√≥n - {columna_numerica}')
    ax1.set_xlabel('Valor')
    ax1.set_ylabel('Frecuencia')
    ax1.legend()
    
    # Boxplot
    ax2.boxplot(valores)
    ax2.set_title(f'Boxplot - {columna_numerica}')
    ax2.set_ylabel('Valor')
    
    # Q-Q plot para normalidad
    stats.probplot(valores, dist="norm", plot=ax3)
    ax3.set_title(f'Q-Q Plot - {columna_numerica}')
    
    plt.tight_layout()
    plt.show()
    
    # Mostrar outliers extremos
    if len(outliers_iqr) > 0:
        print(f"\nüîç Top 5 outliers m√°s extremos (IQR):")
        outliers_extremos = outliers_iqr.nlargest(5)
        for idx, valor in outliers_extremos.items():
            municipio = df.loc[idx, 'municipio_limpio'] if 'municipio_limpio' in df.columns else 'N/A'
            departamento = df.loc[idx, 'departamento_limpio'] if 'departamento_limpio' in df.columns else 'N/A'
            print(f"   {valor:.2f} - {municipio}, {departamento}")
    
    return {
        'outliers_iqr': outliers_iqr,
        'outliers_zscore': outliers_zscore,
        'limites_iqr': (limite_inferior, limite_superior)
    }

# Ejecutar detecci√≥n de outliers si hay columna de √°rea
if df_municipios is not None:
    if 'area_km2' in df_municipios.columns:
        outliers_info = detectar_outliers(df_municipios, 'area_km2')
    elif 'MPIO_CAREA' in df_municipios.columns:
        # Convertir a num√©rico si no lo es
        df_municipios['MPIO_CAREA'] = pd.to_numeric(df_municipios['MPIO_CAREA'], errors='coerce')
        outliers_info = detectar_outliers(df_municipios, 'MPIO_CAREA')
    else:
        print("‚ö†Ô∏è No se encontr√≥ columna de √°rea para an√°lisis de outliers")
```

```python

# CORRECCI√ìN EN CELDA 8: Detecci√≥n de outliers mejorada
def detectar_outliers_corregido(df, columna_numerica=None):
    """
    Detecta outliers - versi√≥n corregida que busca autom√°ticamente columnas de √°rea
    """
    # Si no se especifica columna, buscar autom√°ticamente
    if columna_numerica is None:
        posibles_areas = ['area_km2', 'MPIO_CAREA', 'AREA', 'AREA_KM2']
        columna_encontrada = None
        
        for col in posibles_areas:
            if col in df.columns:
                # Verificar si tiene valores num√©ricos v√°lidos
                test_numeric = pd.to_numeric(df[col], errors='coerce')
                if test_numeric.notna().sum() > 0:
                    columna_encontrada = col
                    break
        
        if columna_encontrada is None:
            print("‚ö†Ô∏è No se encontr√≥ columna de √°rea v√°lida para an√°lisis de outliers")
            print("üìã Columnas disponibles:")
            print(list(df.columns))
            return None
        else:
            columna_numerica = columna_encontrada
            print(f"‚úÖ Usando columna '{columna_numerica}' para an√°lisis de outliers")
    
    # Convertir a num√©rico si no lo es
    if columna_numerica in df.columns:
        df[columna_numerica] = pd.to_numeric(df[columna_numerica], errors='coerce')
    
    if columna_numerica not in df.columns:
        print(f"‚ùå Columna {columna_numerica} no encontrada")
        return None
    
    print(f"\nüìê DETECCI√ìN DE OUTLIERS - {columna_numerica.upper()}")
    print("-" * 45)
    
    # Filtrar valores no nulos
    valores = df[columna_numerica].dropna()
    
    if len(valores) == 0:
        print("‚ùå No hay valores num√©ricos v√°lidos para analizar")
        return None
    
    # M√©todo IQR
    Q1 = valores.quantile(0.25)
    Q3 = valores.quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    outliers_iqr = valores[(valores < limite_inferior) | (valores > limite_superior)]
    
    # M√©todo Z-score
    z_scores = np.abs((valores - valores.mean()) / valores.std())
    outliers_zscore = valores[z_scores > 3]
    
    print(f"üìä Estad√≠sticas b√°sicas:")
    print(f"   Media: {valores.mean():.2f}")
    print(f"   Mediana: {valores.median():.2f}")
    print(f"   Desviaci√≥n est√°ndar: {valores.std():.2f}")
    print(f"   Rango: {valores.min():.2f} - {valores.max():.2f}")
    
    print(f"\nüéØ Detecci√≥n de outliers:")
    print(f"   M√©todo IQR (Q1-1.5*IQR, Q3+1.5*IQR): {len(outliers_iqr)} outliers")
    print(f"   L√≠mites IQR: [{limite_inferior:.2f}, {limite_superior:.2f}]")
    print(f"   M√©todo Z-score (|z| > 3): {len(outliers_zscore)} outliers")
    
    # Visualizaci√≥n
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
    
    # Histograma
    ax1.hist(valores, bins=30, edgecolor='black', alpha=0.7)
    ax1.axvline(valores.mean(), color='red', linestyle='--', label=f'Media: {valores.mean():.2f}')
    ax1.axvline(valores.median(), color='green', linestyle='--', label=f'Mediana: {valores.median():.2f}')
    ax1.set_title(f'Distribuci√≥n - {columna_numerica}')
    ax1.set_xlabel('Valor')
    ax1.set_ylabel('Frecuencia')
    ax1.legend()
    
    # Boxplot
    ax2.boxplot(valores)
    ax2.set_title(f'Boxplot - {columna_numerica}')
    ax2.set_ylabel('Valor')
    
    # Q-Q plot para normalidad
    from scipy import stats
    stats.probplot(valores, dist="norm", plot=ax3)
    ax3.set_title(f'Q-Q Plot - {columna_numerica}')
    
    plt.tight_layout()
    plt.show()
    
    # Mostrar outliers extremos
    if len(outliers_iqr) > 0:
        print(f"\nüîç Top 5 outliers m√°s extremos (IQR):")
        outliers_extremos = outliers_iqr.nlargest(5)
        for idx, valor in outliers_extremos.items():
            municipio = df.loc[idx, 'municipio'] if 'municipio' in df.columns else 'N/A'
            departamento = df.loc[idx, 'departamento'] if 'departamento' in df.columns else 'N/A'
            print(f"   {valor:.2f} - {municipio}, {departamento}")
    
    return {
        'outliers_iqr': outliers_iqr,
        'outliers_zscore': outliers_zscore,
        'limites_iqr': (limite_inferior, limite_superior)
    }
```

```python
def preparar_para_integracion(df):
    """
    Prepara el DataFrame de municipios para integraci√≥n con datos de cobertura
    """
    print(f"\nüîó PREPARANDO PARA INTEGRACI√ìN")
    print("-" * 35)
    
    # Seleccionar columnas esenciales
    columnas_esenciales = ['codigo_municipio']
    
    if 'municipio_limpio' in df.columns:
        columnas_esenciales.append('municipio_limpio')
        
    if 'departamento_limpio' in df.columns:
        columnas_esenciales.append('departamento_limpio')
        
    if 'area_km2' in df.columns:
        columnas_esenciales.append('area_km2')
    
    # Agregar columnas originales importantes
    columnas_originales = ['MPIO_CNMBR', 'DPTO_CNMBR', 'MPIO_CAREA']
    for col in columnas_originales:
        if col in df.columns:
            columnas_esenciales.append(col)
    
    # Crear DataFrame final
    df_final = df[columnas_esenciales].copy()
    
    # Renombrar para consistencia
    rename_dict = {
        'municipio_limpio': 'municipio',
        'departamento_limpio': 'departamento'
    }
    df_final = df_final.rename(columns=rename_dict)
    
    print(f"‚úÖ DataFrame preparado para integraci√≥n:")
    print(f"   Registros: {len(df_final)}")
    print(f"   Columnas: {list(df_final.columns)}")
    
    return df_final

# Preparar datos para an√°lisis
if df_municipios is not None:
    df_municipios_final = preparar_para_integracion(df_municipios)
    
    print("\nüìã Muestra de datos preparados:")
    display(df_municipios_final.head())
```

```python
def analisis_estadistico_profundo(df):
    """
    An√°lisis estad√≠stico m√°s profundo incluyendo pruebas de normalidad
    """
    print(f"\nüî¨ AN√ÅLISIS ESTAD√çSTICO PROFUNDO")
    print("=" * 35)
    
    # Columnas num√©ricas para an√°lisis
    columnas_numericas = df.select_dtypes(include=[np.number]).columns
    print(f"Columnas num√©ricas encontradas: {list(columnas_numericas)}")
    
    for col in columnas_numericas:
        if col in df.columns:
            valores = df[col].dropna()
            if len(valores) > 0:
                print(f"\nüìê AN√ÅLISIS DE: {col.upper()}")
                print("-" * 30)
                
                # Estad√≠sticas b√°sicas
                print(f"Estad√≠sticas descriptivas:")
                print(f"  N: {len(valores)}")
                print(f"  Media: {valores.mean():.4f}")
                print(f"  Mediana: {valores.median():.4f}")
                print(f"  Moda: {valores.mode().iloc[0] if len(valores.mode()) > 0 else 'N/A'}")
                print(f"  Desv. est√°ndar: {valores.std():.4f}")
                print(f"  Varianza: {valores.var():.4f}")
                print(f"  Asimetr√≠a (skewness): {valores.skew():.4f}")
                print(f"  Curtosis: {valores.kurtosis():.4f}")
                
                # Interpretaci√≥n de asimetr√≠a
                skew_val = valores.skew()
                if abs(skew_val) < 0.5:
                    skew_interp = "aproximadamente sim√©trica"
                elif skew_val > 0.5:
                    skew_interp = "asim√©trica positiva (cola derecha)"
                else:
                    skew_interp = "asim√©trica negativa (cola izquierda)"
                print(f"  Interpretaci√≥n: Distribuci√≥n {skew_interp}")
                
                # Prueba de normalidad
                if len(valores) >= 8:  # M√≠nimo para Shapiro-Wilk
                    if len(valores) <= 5000:  # Shapiro-Wilk para muestras peque√±as
                        stat, p_value = stats.shapiro(valores)
                        test_name = "Shapiro-Wilk"
                    else:  # Kolmogorov-Smirnov para muestras grandes
                        stat, p_value = stats.normaltest(valores)
                        test_name = "D'Agostino"
                    
                    print(f"\nüß™ Prueba de normalidad ({test_name}):")
                    print(f"  Estad√≠stico: {stat:.6f}")
                    print(f"  p-valor: {p_value:.6f}")
                    
                    if p_value > 0.05:
                        print(f"  ‚úÖ Los datos parecen seguir una distribuci√≥n normal (p > 0.05)")
                    else:
                        print(f"  ‚ùå Los datos NO siguen una distribuci√≥n normal (p ‚â§ 0.05)")

# Ejecutar an√°lisis estad√≠stico profundo
if df_municipios_final is not None:
    analisis_estadistico_profundo(df_municipios_final)
```

```python
def analisis_estadistico_profundo_corregido(df):
    """
    An√°lisis estad√≠stico corregido que identifica y convierte columnas num√©ricas
    """
    print(f"\nüî¨ AN√ÅLISIS ESTAD√çSTICO PROFUNDO")
    print("=" * 35)
    
    # Identificar columnas que pueden ser num√©ricas
    columnas_posiblemente_numericas = []
    
    for col in df.columns:
        if col != 'codigo_municipio':  # Excluir c√≥digos
            # Intentar convertir a num√©rico
            test_numeric = pd.to_numeric(df[col], errors='coerce')
            valores_numericos = test_numeric.notna().sum()
            
            if valores_numericos > 0:
                # Si m√°s del 50% son num√©ricos, considerarla num√©rica
                if valores_numericos / len(df) > 0.5:
                    columnas_posiblemente_numericas.append(col)
                    # Convertir la columna
                    df[col] = test_numeric
    
    print(f"Columnas identificadas como num√©ricas: {columnas_posiblemente_numericas}")
    
    if len(columnas_posiblemente_numericas) == 0:
        print("‚ö†Ô∏è No se encontraron columnas num√©ricas para an√°lisis estad√≠stico")
        print("üìã Columnas disponibles en el dataset:")
        for i, col in enumerate(df.columns, 1):
            muestra = df[col].iloc[0] if len(df) > 0 else 'N/A'
            print(f"   {i}. {col}: {type(muestra).__name__} (ejemplo: {muestra})")
        return
    
    # An√°lizar cada columna num√©rica
    for col in columnas_posiblemente_numericas:
        valores = df[col].dropna()
        if len(valores) > 0:
            print(f"\nüìê AN√ÅLISIS DE: {col.upper()}")
            print("-" * 30)
            
            # Estad√≠sticas b√°sicas
            print(f"Estad√≠sticas descriptivas:")
            print(f"  N: {len(valores)}")
            print(f"  Media: {valores.mean():.4f}")
            print(f"  Mediana: {valores.median():.4f}")
            print(f"  Desv. est√°ndar: {valores.std():.4f}")
            print(f"  Varianza: {valores.var():.4f}")
            print(f"  Asimetr√≠a (skewness): {valores.skew():.4f}")
            print(f"  Curtosis: {valores.kurtosis():.4f}")
            
            # Interpretaci√≥n de asimetr√≠a
            skew_val = valores.skew()
            if abs(skew_val) < 0.5:
                skew_interp = "aproximadamente sim√©trica"
            elif skew_val > 0.5:
                skew_interp = "asim√©trica positiva (cola derecha)"
            else:
                skew_interp = "asim√©trica negativa (cola izquierda)"
            print(f"  Interpretaci√≥n: Distribuci√≥n {skew_interp}")
            
            # Prueba de normalidad
            from scipy import stats
            if len(valores) >= 8:
                if len(valores) <= 5000:
                    stat, p_value = stats.shapiro(valores)
                    test_name = "Shapiro-Wilk"
                else:
                    stat, p_value = stats.normaltest(valores)
                    test_name = "D'Agostino"
                
                print(f"\nüß™ Prueba de normalidad ({test_name}):")
                print(f"  Estad√≠stico: {stat:.6f}")
                print(f"  p-valor: {p_value:.6f}")
                
                if p_value > 0.05:
                    print(f"  ‚úÖ Los datos parecen seguir una distribuci√≥n normal (p > 0.05)")
                else:
                    print(f"  ‚ùå Los datos NO siguen una distribuci√≥n normal (p ‚â§ 0.05)")

outliers_info = detectar_outliers_corregido(df_municipios_final)
                    
```

```python
def analizar_y_resolver_duplicados(df):
    """
    Analiza y resuelve duplicados en c√≥digos DIVIPOLA
    """
    print(f"\nüîç AN√ÅLISIS DETALLADO DE DUPLICADOS")
    print("-" * 45)
    
    # Identificar duplicados
    duplicados_mask = df['codigo_municipio'].duplicated(keep=False)
    df_duplicados = df[duplicados_mask].sort_values('codigo_municipio')
    
    print(f"Total de registros duplicados: {len(df_duplicados)}")
    
    if len(df_duplicados) > 0:
        print(f"C√≥digos √∫nicos duplicados: {df_duplicados['codigo_municipio'].nunique()}")
        
        # Mostrar ejemplos de duplicados
        print(f"\nüìã Ejemplos de duplicados:")
        for codigo in df_duplicados['codigo_municipio'].unique()[:5]:
            ejemplos = df_duplicados[df_duplicados['codigo_municipio'] == codigo]
            print(f"\nC√≥digo {codigo}:")
            display(ejemplos[['codigo_municipio', 'municipio', 'departamento']])
        
        # Estrategias de resoluci√≥n
        print(f"\nüõ†Ô∏è ESTRATEGIAS DE RESOLUCI√ìN:")
        print("1. Mantener primer registro (keep='first')")
        print("2. Mantener √∫ltimo registro (keep='last')")
        print("3. An√°lisis inteligente - mantener registro m√°s completo")
        
        # Opci√≥n 3: An√°lisis m√°s inteligente
        # Priorizar registros con m√°s informaci√≥n completa
        campos_relevantes = ['municipio', 'departamento']
        if 'area_km2' in df.columns:
            campos_relevantes.append('area_km2')
        elif 'MPIO_CAREA' in df.columns:
            campos_relevantes.append('MPIO_CAREA')
        
        df_temp = df.copy()
        df_temp['completitud'] = df_temp[campos_relevantes].notna().sum(axis=1)
        
        # Mantener el registro m√°s completo para cada c√≥digo
        df_sin_duplicados = (df_temp.sort_values(['codigo_municipio', 'completitud'], ascending=[True, False])
                            .drop_duplicates(subset='codigo_municipio', keep='first')
                            .drop(columns=['completitud']))
        
        print(f"\nüìä RESULTADOS DE LIMPIEZA:")
        print(f"Original: {len(df)} registros")
        print(f"Sin duplicados: {len(df_sin_duplicados)} registros")
        print(f"Registros removidos: {len(df) - len(df_sin_duplicados)}")
        
        # Verificar que ya no hay duplicados
        verificacion = df_sin_duplicados['codigo_municipio'].duplicated().sum()
        print(f"‚úÖ Duplicados restantes: {verificacion}")
        
        return df_sin_duplicados
    
    else:
        print("‚úÖ No se encontraron duplicados en los c√≥digos DIVIPOLA")
        return df

# Aplicar resoluci√≥n de duplicados
if df_municipios_final is not None:
    df_municipios_limpio = analizar_y_resolver_duplicados(df_municipios_final)
```
